```python
# import main libraries
from app_store_scraper import AppStore
from pprint import pprint
```

```python
key_word = "Baedal Minjok" # insert keyword
country = "kr" # select country code https://en.wikipedia.org/wiki/ISO_3166-1_alpha-2
count = 20 # data count needed

app = AppStore(country=country, app_name=key_word)
app.review(how_many=count)

pprint(app.reviews)
pprint(app.reviews_count)
```

```python
import pandas as pd

df = pd.DataFrame(app.reviews)
df
```

```python
df.to_csv('Baedal Minjok.csv', header=True, index=False)
```
